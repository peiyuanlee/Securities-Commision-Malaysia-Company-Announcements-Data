```python
#cleaning category data
cat = pd.read_csv(r"C:\Users\pylee\Documents\table-Copy of sample_CA_data (for manual categorization).csv")
cat = cat[1:]
cat.rename(columns = {'a. A list of announcements that CS wishes to be alerted to':'list', "Unnamed: 2": "subcategory", "Unnamed: 3": "keywords"}, inplace = True)


def prepare_table(col): 
    """
    Removes NaN values and fills in empty spaces
    """
    num = 1
    for i in cat[col]:
        if(pd.isna(i)):
            cat.loc[num][col] = cat.loc[num-1][col]
        num+=1
    
       
prepare_table("list")
prepare_table("Bursa Category")
prepare_table("keywords")

num = 1
for i in cat["list"]:
    cat["list"].loc[num] = i[3:]
    num +=1
    
#clean data
cat["Bursa Category"] = cat["Bursa Category"].replace({"General announcements": "General Announcement", "Unusual Market Activity (UMA)": "Unusual Market Activity",
                                                       "Reply to query": "Reply to Query"})
cat["subcategory"] = cat["subcategory"].replace({"Practice Note 17/ Guidance Note 3": "Practice Note 17 / Guidance Note 3", 
                            "New Issue of Securities (Chapter 6 of Listing Requirements)": "New Issue of Securities (Chapter 6 of listing requirements)",
                            "Others": "OTHERS"})
cat["keywords"] = cat["keywords"].replace({"Delay in Issuance of \nDelay in release of first quarter financial report\nDelay in release the (unaudited/ audited) quarterly report\nDelay in release of Annual Report\nDelay in release the audited financial statements": "Delay",
                                          "(Suspension/ Leave) (or from duties)of directors/ principal officer/ CEO": "Suspension Leave from duties", 
                                         "removal/ appointment of directors/ Requisition by shareholders": "removal appointment Requisition", 
                                         "Resignation of auditors\nnomination of auditors\nchange of auditors\nchange of external auditors\nappointment of auditors": "auditors",
                                         "acquisition, disposal": "acquisition disposal", 
                                           "bonus issue, private placement, fund raising, share issuance, esos": "bonus issue private placement fund raising share issuance esos",
                                          "Deviation of 10% or more between the announced unaudited financial statements and the audited financial statements": "deviation",
                                          "Award of contracts/ Collaborative Agreement" : "Award of contracts Collaborative Agreement",
                                          "Special notice for removal/ appointment of directors/ Requisition by shareholders": "Requisition directors director shareholders removal appointment"})

#cat = cat[cat["subcategory"]!= "Others"]
#cat = cat[cat["keywords"]== "-"]
#data = data[(data["sub_category"]!= "OTHERS") & (data["sub_category"]!= "Multiple Proposals")]

cat
```

```python
import nltk
from nltk.corpus import stopwords

stopword = stopwords.words("english")
stopword.extend("-")
label = []
num = 0
while num != len(data):
    labels = cat[cat["Bursa Category"] == data["assigned_category"].loc[num]]
    if(len(labels) == 1):
        label.append(str(labels["list"]))
    else:
        labels = labels[labels["subcategory"] == data["sub_category"].loc[num]]
        if(len(labels) == 1):
            label.append(str(labels["list"]))
        else:
            count = 0
            labLen = len(label)
            for keyword in labels["keywords"]:
                i = keyword.split()
                words = [word for word in i if word not in stopword]
                for word in words:
                    if(word.lower() in data["title"].loc[num].lower()):
                        l = labels.iloc[count]
                        label.append(str(l["list"]))
                        break
                if(labLen != len(label)):
                    break
                count += 1   
            if((count == len(labels)) & (data["sub_category"].loc[num] == "OTHERS")):
                label.append("Clarification of announcement/ UMA")
            elif(count == len(labels)):
                label.append("")
    num+=1
label
```
